```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\DELL\Downloads\Placement_Data_Full_Class (1).csv")
df.head()
```

```python
df = df[["Mba %", "salary"]]
df.head()
```

```python
df.shape
```

```python
plt.scatter(df['Mba %'],df['salary'])
plt.xlabel('Mba(%)')
plt.ylabel('Salary')
```

```python
df = df.dropna(subset=["salary"])  # Drops rows where salary is NaN
```

```python
X = df[["Mba %"]]    # Features should be a 2D array (that's why it's double brackets)
y = df["salary"]     # Target/label
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
y_pred=lr.predict(X_test)
```

```python
plt.scatter(df['Mba %'],df['salary'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('Mba %')
plt.ylabel('Salary')
```
